**Merge authorInfo.csv and article.csv**

```python
import pandas as pd
import numpy as np
```

```python
authorInfo_df = pd.read_csv('authorInfo.csv')
articleInfo_df = pd.read_csv('articleInfo.csv')
join_df = authorInfo_df.merge(articleInfo_df, how = 'outer', on = 'Article No.').fillna(0)
join_df.to_csv('merge.csv')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

**1. Plot the yearly_publication figure, in which the x-axis is the year, the y-axis is
the number of articles published during that year.**

```python
df = join_df
df = df.sort_values('Year', ascending = True)
x = df['Year'].drop_duplicates(keep = 'first', inplace = False)
y = df.Year.value_counts()[df.Year.sort_values()].drop_duplicates()
plt.bar(x, y)
plt.title("yearly_publication") # set the title of the figure
plt.xlabel("Year") # x-axis label
plt.ylabel("Article No.") # y-axis label
plt.show()
```

**2. Plot the yearly_citation figure, in which the x-axis is the year, the y-axis is the
total number of citations during that year.**

```python
grouped_data = df.groupby('Year')['Citation'].sum()
grouped_data.plot(x = 'Year', y = 'Citation', kind = 'bar')
plt.xlabel("Year")
plt.ylabel("Number of Citations")
plt.title("Yearly Citations")
```
